```python
import pandas as pd

# Replace 'RiseDinnerSheet.csv' with the actual file path if it's not in the same directory as your script.
csv_file_path = 'Rise Dinner Sheet - Week 7.csv'

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(csv_file_path)

# Print the DataFrame
print(df)

```

```python
BrockCost = 0
ErikCost = 0
EthanCost = 0
ColbyCost = 0
for row in df.itertuples():
    ey = False
    ej = False
    cm = False
    bg = False
    
    counter = 0
    
    buyer = row.Buyer
    
        
    
    if row.Communal == 'x':
        c = row.Cost / 4
        BrockCost += c
        ErikCost += c
        EthanCost += c
        ColbyCost += c
        counter = 4
        
    
    if row.Erik == 'x':
        ej = True
        counter += 1
    if row.Brock == 'x':
        bg = True
        counter += 1
    if row.Colby == 'x':
        cm = True
        counter += 1
    if row.Ethan == 'x':
        ey = True
        counter += 1
    
    
    if ej:
        ErikCost += row.Cost / counter  
    if ey:
        EthanCost += row.Cost / counter
    if bg:
        BrockCost += row.Cost / counter
    if cm:
        ColbyCost += row.Cost / counter
        
    if row.Buyer[:4] == "Erik":
        ErikCost -= row.Cost
    if row.Buyer[:5] == "Brock":
        BrockCost -= row.Cost   
    if row.Buyer[:5] == "Ethan":
        EthanCost -= row.Cost
    if row.Buyer[:5] == "Colby":
        ColbyCost -= row.Cost
        
total_cost = df['Cost'].sum()
print("The total cost this week was: $",total_cost)
        
    
        
print("Erik owes(+) / isOwed(-): $",ErikCost)
print("Brock owes(+) / isOwed(-): $",BrockCost)
print("Ethan owes(+) / isOwed(-): $",EthanCost)
print("Colby owes(+) / isOwed(-): $",ColbyCost)
```
